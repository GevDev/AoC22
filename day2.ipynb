```python
def find_ending(me, opponent):
    score_matrix = [[4,8,3],
                    [1,5,9],
                    [7,2,6]]
    selector = {
        "X": 0,
        "Y": 1,
        "Z": 2,
        "A": 0,
        "B": 1,
        "C": 2
    }

    if me == "X":
        return min(score_matrix[selector[opponent]])
    if me == "Y":
        return sorted(score_matrix[selector[opponent]])[1]
    if me == "Z":
        return max(score_matrix[selector[opponent]])


total_score2 = 0
for entry in input2:
    half = entry.split(" ")
    total_score2 += find_ending(half[1], half[0])

print(total_score2)
```
