```python
def get_priority(char):
    if ord(char) >= 65 and ord(char) <= 90:
        return ord(char) - 38
    elif ord(char) >= 97 and ord(char) <= 122:
        return ord(char) - 96


def find_dup_char(word):
    first_half = word[:len(word)//2]
    second_half = word[len(word)//2:]

    # Python makes this too easy, hashmap is probably the way to go about in other languages
    for char in first_half:
        if char in second_half:
            return get_priority(char)
```

```python
total_priority = 0
for entry in input2:
    total_priority += find_dup_char(entry)

print(total_priority)
```

```python
# Frisk the elf sacks
def find_badge(elf1_sack, elf2_sack, elf3_sack):
    for char in elf1_sack:
        if char in elf2_sack and char in elf3_sack:
            return get_priority(char)

total_priority_2 = 0
elf_sack = 0
while elf_sack+2 < len(input4):
    total_priority_2 += find_badge(input4[elf_sack], input4[elf_sack+1], input4[elf_sack+2])
    elf_sack += 3


print(total_priority_2)
```

```python
import string

# Part one written by openai chatbot
# the priorities of the different item types
priorities = {c: i+1 for i, c in enumerate(string.ascii_lowercase)}
priorities.update({c: i+1+26 for i, c in enumerate(string.ascii_uppercase)})

# the list of rucksacks
rucksacks = input2
# [
#     "vJrwpWtwJgWrhcsFMMfFFhFp",
#     "jqHRNqRjqzjGDLGLrsFMfFZSrLrFZsSL",
#     "PmmdzqPrVvPwwTWBwg",
#     "wMqvLMZHhHMvwLHjbvcjnnSBnvTQFn",
#     "ttgJtRGJQctTZtZT",
#     "CrZsJsPPZsGzwwsLwLmpwMDw",
# ]

# the sum of priorities of items that appear in both compartments of each rucksack
total_priority = 0

# find the item that appears in both compartments of each rucksack
for rucksack in rucksacks:
    # split the rucksack into the two compartments
    comp1, comp2 = rucksack[:len(rucksack)//2], rucksack[len(rucksack)//2:]
    # find the common items in the two compartments
    common_items = set(comp1) & set(comp2)
    # add the priority of the common item to the total
    if common_items:
        total_priority += priorities[common_items.pop()]

# print the total priority
print(total_priority)
```
