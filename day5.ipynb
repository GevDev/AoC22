```python
char_index = 0
stacks = defaultdict(lambda: [])
counter = 1
while input1[char_index+1] != "1":
    if input1[char_index] == "[":
        stacks[counter].insert(0, input1[char_index+1])
    char_index += 4
    counter += 1
    if input1[char_index-1] == "\n":
        counter = 1

def parse_instructions(line):
    split_line = line.split(" ")
    return int(split_line[1]), int(split_line[3]), int(split_line[5])
```

```python
# Part 1
lines = input1.split("\n\n")[1].split("\n")
for line in lines:
    instruct = parse_instructions(line)
    for x in range(0, instruct[0]):
        # print(stacks[instruct[1]].reverse())
        stacks[instruct[2]].append(stacks[instruct[1]].pop())

for entry in stacks:
    print(entry, stacks[entry][-1])
```

```python
# Part 2
lines = input1.split("\n\n")[1].split("\n")
for line in lines:
    instruct = parse_instructions(line)
    # print(stacks[instruct[1]][0:len(instruct[1])])
    removed_stacks = stacks[instruct[1]][len(stacks[instruct[1]])-instruct[0]:len(stacks[instruct[1]])]
    stacks[instruct[1]] = stacks[instruct[1]][0:len(stacks[instruct[1]])-instruct[0]]
    stacks[instruct[2]] += removed_stacks


for entry in stacks:
    print(entry, stacks[entry][-1])
```
